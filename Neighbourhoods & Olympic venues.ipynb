```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup as bs

url = 'https://www.2020games.metro.tokyo.lg.jp/eng/taikaijyunbi/taikai/map/detail_map/index.html#tabOlympic'
response = requests.get(url)
response
```

```python
soup = bs(response.text, "html.parser")
soup.findAll('a')
time.sleep(1)
```

```python
for link in soup.find_all("a"):
    print("Inner Text: {}".format(link.text))
```

```python
import pandas as pd
import re as re 

data = []
row1 = [link.text.strip() for link in soup.find_all("a")]
row2 = [link.text.strip() for link in soup.find_all("span", attrs={'class':'icon__text'})]
data.append([row1,row2])

olympics_df = pd.DataFrame(data, columns=["Venue", "Sports"])
print(olympics_df)
```

```python
url = 'https://www.jrailpass.com/blog/tokyo-2020-olympics/venues'
response_new = requests.get('https://www.jrailpass.com/blog/tokyo-2020-olympics/venues')
response_new

soup_new = bs(response_new.text, "html.parser")
table = soup_new.find_all('table')[0]
df = pd.read_html(str(table))
df
```

```python
df = pd.DataFrame(df[1:],columns=df[0])
df
```

```python
# Compute the Value column with `agg`.
data = df.astype(str).agg(','.join)
data
```

```python
df_new = pd.DataFrame(data)
df_new
```

```python
olympics_table = soup_new.find("table")
olympics_data = olympics_table.tbody.find_all("tr")
olympics_data
```

```python
headings = []
for td in olympics_data[0].find_all("td"):
    # remove any newlines and extra spaces from left and right
    headings.append(td.strong.text.replace('\n', ' ').strip())

print(headings)
```
